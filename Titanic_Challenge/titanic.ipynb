```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
X = pd.read_csv("./Dataset/train.csv")
```

```python
X.head()
```

```python
y = X.pop("Survived")
y.head()
```

```python
X.describe()
```

```python
X['Age'].fillna(X.Age.mean(),inplace=True)
X.describe()
```

```python
#just to get the numeric values out of data
numeric_variables = list(X.dtypes[X.dtypes!="object"].index)
X[numeric_variables].head()
```

```python
model = RandomForestRegressor(n_estimators=100,oob_score=True,random_state=42)
model.fit(X[numeric_variables],y)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model.oob_score_
```

```python
y_oob = model.oob_prediction_
print("c-stat or auc",roc_auc_score(y,y_oob))
```

```python
#y_oob is out of bag prediction
#y_oob gives probability of survival of every person
y_oob
```

```python
def describe_categorical(X):
    from IPython.display import display, HTML
    display(HTML(X[X.columns[X.dtypes=='object']].describe().to_html()))
```

```python
describe_categorical(X)
```

```python
X.drop(['Name','Ticket','PassengerId'],axis=1,inplace=True)
```

```python
#change the first cabin variable to be only the first leter or None
def clean_cabin(x):
    try:
        return x[0]
    except:
        return "None"
X['Cabin']=X.Cabin.apply(clean_cabin)
```

```python
X.head()
```

```python
categorical_variables = ['Sex','Cabin','Embarked']
```

```python
for variable in categorical_variables:
    X[variable].fillna("Missing",inplace=True)
    #create array of dummies
    dummies = pd.get_dummies(X[variable],prefix=variable)
    X = pd.concat([X,dummies],axis=1)
    X.drop([variable],axis=1,inplace=True)
```

```python
X.head()
```

```python
#to look at all columns some might be hidden due to lack of spac
def printall(X,max_rows=10):
    from IPython.display import display, HTML
    display(HTML(X.to_html(max_rows=max_rows)))
printall(X)
```

```python
model = RandomForestRegressor(100,oob_score=True,n_jobs=-1,random_state=42)
model.fit(X,y)
print("C- State or AUC",roc_auc_score(y,model.oob_prediction_))
```

```python
model.feature_importances_
# ^ Not very helpful lol :P
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
feature_importances = pd.Series(model.feature_importances_,index=X.columns)
feature_importances.sort()
feature_importances.plot(kind='barh',figsize=(7,6))
```
